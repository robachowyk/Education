```python
""" Dans python les caracteres sont en UTF8, 
    on cree des fonctions pour reconnaitre si un caractere est une minuscule ou une majuscule """

def minuscule(caractere):
    if (caractere >= 'a') and (caractere <= 'z'):
        return True
    else:
        return False
    
def majuscule(caractere):
    if (caractere >= 'A') and (caractere <= 'Z'):
        return True
    else:
        return False
```

```python
""" Fonction qui charge un fichier, et qui le transforme en liste de mots (attention  la ponctuation) """

def text_to_list():
    mon_fichier = open("Wonderland.txt", "r")
    contenu = mon_fichier.read() # contenu est le texte integral
    
    #on garde seulement les caracteres de l alphabet
    for caractere in contenu:
        if minuscule(caractere)==False and majuscule(caractere)==False:
            contenu=contenu.replace(caractere,' ')
    texte=contenu.split()
    return texte

print(text_to_list())
```

```python
""" Fonction qui compte le nombre doccurence de chaque mot et qui le renvoie sous la forme dun dictionnaire """

def comptage(liste):
    resultat={}
    for mot in liste:
        resultat[mot]=resultat.get(mot,0)+1  
    return resultat

comptage(text_to_list())
```

```python
""" Fonction qui renvoie les mots les plus frquents """

import pandas as pds

def mots_frequents(n,dico):
    S=pds.Series(dico)
    return S.nlargest(n).index.to_list()

mots_frequents(10,comptage(text_to_list()))
```

```python
""" Meme analyse avec le nouveau fichier
    le fichier Wonderland3 est le fichier du prof modifi (delimiter='\t' probleme de tabulation) 
    On met dans un dataframe le fichier modifi, 
    la premiere colonne : mot indique le mot du texte,
    la seconde colonne : type indique le type du mot, 
    la 3e colonne : ref indique le mot de reference au on attribue au mot du texte """

data = pds.read_csv('Wonderland3.txt', delimiter="\t", header=None)
data.columns=['mot','type','ref']
data
```

```python
""" La seconde colonne : type indique le type du mot,
    on regarde la liste des types qui existent """

""" On peut alors donner une liste de type qui ne sont pas interessants comme critere 
    pour creer plus tard la fonction qui renverra les mots les plus pertinents 
    ["TO",",","''","SENT","IN","DT","CC",":","RB"] """

def liste_des_types():
    return data['type'].value_counts()

liste_des_types()

```

```python
""" On modifie le dataframe 'data' pour combler les ref '<unknown>' """

def modification_data():
    for ligne in data.index:
        if data['ref'][ligne]=='<unknown>':
            mot_a_transformer=str(data['mot'][ligne])
            for caractere in mot_a_transformer:
                if minuscule(caractere)==False and majuscule(caractere)==False:
                    #si le mot comporte d'autres elements que des lettres on les supprime
                    mot_a_transformer=mot_a_transformer.replace(caractere,'')
                if majuscule(caractere)==True:
                    #si le mot est en majuscule on le met en minuscule
                    mot_a_transformer=mot_a_transformer.lower()
                if mot_a_transformer[-1:]=='s': 
                    #si le mot est au pluriel on le met au singulier
                    mot_a_transformer=mot_a_transformer.replace(mot_a_transformer[-1:],'')
            new_mot=mot_a_transformer
            data['ref'][ligne]=new_mot
    return data
            
modification_data()
```

```python
""" Fonction qui compte le nombre doccurence de chaque mot """

def comptage_des_ref():
    return data['ref'].value_counts()   
    
comptage_des_ref()
```

```python
""" Fonction qui renvoie les mots les plus pertinents """

""" On retourne la liste des n mots les plus frequents dont le type n'est pas parmi les types
    designe plus haut comme n'etant pas interessant 
    On utilise ~ isin qui signifie : n'est pas dans la liste suivante """

def mots_plus_pertinents(n):
    occurrence=data[~data['type'].isin(["TO",",","''","SENT","IN","DT","CC",":","RB"])]['ref'].value_counts()
    return occurrence.nlargest(n).index.to_list()

mots_plus_pertinents(20)
```

```python
""" Fonction bigrammes : pour repertorier toutes les squences de 2 mots intressantes (les n plus frequentes) """
""" On garde la ponctuation pour plus tard ecrire des phrases coherentes avec de la ponctuation """

def bigrammes(n):
    resultat={}
    liste=data['ref'].to_list()
    for i in range(len(liste)-1):
        resultat[tuple([liste[i],liste[i+1]])]=resultat.get(tuple([liste[i],liste[i+1]]),0)+1 
    S=pds.Series(resultat)
    return S.nlargest(n).index.to_list()

bigrammes(20)
```

```python
""" generalisation pour repertorier les squences de longueur arbitraire (a indice_coherence elements) 
    plus indice_coherence est grand, plus le texte final qu on cherche a generer sera coherent et moins il sera original """

def n_grammes(indice_coherence):
    resultat={}
    liste=data['ref'].to_list()
    for i in range(len(liste)-1):
        resultat[tuple(liste[i:i+indice_coherence])]=resultat.get(tuple(liste[i:i+indice_coherence]),0)+1 
    return resultat

n_grammes(2)
```

```python
""" On calcule le nbr de cas pour lequel un mot en suit un autre dans le texte """

""" compte_mot_debut compte le nbr de cas ou une sequence de mots apparait au debut d'un n_gramme """

def compte_mot_debut(mots, grammes): 
    # mots est une liste de mots
    cpt=0
    for cle in grammes:
        L=list(cle)
        if L[:len(mots)]==mots:
            cpt+=grammes[cle]
    return cpt

compte_mot_debut(['of'], n_grammes(2))
```

```python
""" probabilte_2_mots calcule la proba que le mot2 apparaisse apres le mot1 dans un n_gramme 
    cette proba vaut :
    nbr de fois ou mot2 apparait juste apres mot1 dans le texte / nbr de fois ou mot1 apparait au debut d'un n_gramme """


def probabilite_2_mots(mot1,mot2):
    # mot1 doit etre une liste (par soucis de coherence avec la fonction compte_mot_debut)
    grammes = n_grammes(2)    
    return grammes[(mot1[0],mot2)]/compte_mot_debut(mot1, grammes)


probabilite_2_mots(['of'],'the')
```

```python
""" Fonction qui pour une sequence de mots retourne le dictionnaire des mots 
    qui la suivent dans le texte avec leur proba de la suivre """

def serie_proba_n_grammes(mots):
    # mots est une liste de mots
    dico={}
    grammes = n_grammes(len(mots)+1)
    p = compte_mot_debut(mots, grammes)
    for element in grammes:
        if list(element[:len(mots)])==mots:
            dico[element[-1]] = grammes[element]/p
    return dico

serie_proba_n_grammes(['of', 'have'])
```

```python
""" Fonction qui retourne les n mots dont la proba est la plus grande de suivre le mot pass en parametre """

def suivant_probable(mots,n):
    # mots est une liste de mots
    return pds.Series(serie_proba_n_grammes(mots)).nlargest(n).index.to_list()

suivant_probable(['of'],3)
```

```python
""" Fonction qui genere un texte de n mots """
""" Apres chaque mot ajout on choisi a l aide de numpy.random.choice un nouveau mot parmi ceux
    dans le dico des mots qui suivent la sequence des quelques mots precedants """

import random
import numpy as np

def liste_to_str(liste):
    s = ""
    for element in liste:
        s += str(element)+" "
    return s

def generer_texte(n,indice_coherence):
    phrase = []
    phrase.append(random.choice(data['mot'].to_list()))
    for i in range(1, n):
        
        choix_mot = serie_proba_n_grammes(phrase[-min(len(phrase),-indice_coherence):])
        
        mot_suivant = np.random.choice(list(choix_mot.keys()), p=list(choix_mot.values()))
        
        phrase.append(mot_suivant)
        
    return liste_to_str(phrase)

generer_texte(40,3)
```
